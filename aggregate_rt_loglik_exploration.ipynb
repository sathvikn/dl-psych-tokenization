```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

```python
sns.set_style('darkgrid')
```

```python
surprisal_rts = pd.read_csv("rt_surprisals.csv")
```

```python
sns.regplot(data = surprisal_rts, x = "coca_5gram_baseline_surprisal", y = "total_rt", scatter_kws={'alpha':0.2}, label = "baseline")
sns.regplot(data = surprisal_rts, x = "coca_5gram_bpe_surprisal", y = "total_rt", scatter_kws={'alpha':0.2}, label = "BPE")
plt.xlabel("sentence_log_likelihood")
plt.ylabel("sentence_rt")
plt.legend()
```

```python
scipy.stats.pearsonr(surprisal_rts['coca_5gram_baseline_surprisal'], surprisal_rts['total_rt'])
```

```python
scipy.stats.pearsonr(surprisal_rts['coca_5gram_bpe_surprisal'], surprisal_rts['total_rt'])
```

```python
from pymer4 import Lmer
```

```python
model = Lmer('total_rt ~ coca_5gram_baseline_surprisal + (1 | word_count)', data=surprisal_rts)
```
